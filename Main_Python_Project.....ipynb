**2,DATA PRE-PROCESSING**

```python
# check missing values

print("/t missing values:")
print(df.isnull().sum())
```

```python
# REMOVING DUPLICATES

# Check duplicate records
print(df.duplicated().sum())
```

```python
# correcting data types
print(df.dtypes)

# convert date to integer
df['date'] = pd.to_datetime(df['date'])

energy_columns= ['solar_energy','wind_energy','other_renewable_energy','total_renewable_energy']

for col in energy_columns: df[col]=pd.to_numeric(df[col],errors='coerce')
```

```python
#Creating deriverd columns

df['total_renewable_energy']= (
    df['solar_energy']+ df['wind_energy']+ df['other_renewable_energy'])

df['solar_percentage']=(df['solar_energy']/df['total_renewable_energy'])*100

df['wind_percentage']= (df['wind_energy']/df['total_renewable_energy'])*100

df['other_renewable_energy_percentage']= (df['other_renewable_energy']/df['total_renewable_energy'])*100

df['growth_rate']=(df.groupby('state_name')['total_renewable_energy'].pct_change()*100 )

df[[ 'state_name','total_renewable_energy','solar_percentage','wind_percentage','other_renewable_energy_percentage']].head()
```

```python
# filtering and aggregatting data

#Filltering data
df['year']= pd.to_datetime(df['date']).dt.year

# spcific date
filtered_data=df[df['date']== '2020-05-28']
print(filtered_data)
```

```python
#filter specific state

Tamil_Nadu_data= df[df['state_name']=='Tamil Nadu']
print(Tamil_Nadu_data.head())
```

```python
# Aggregatting data

#State wise total Renewable energy generation

state_generation= df.groupby('state_name')['total_renewable_energy'].sum()
print(state_generation)
```

```python
#Average generation by state

Avg_gneration=df.groupby('state_name')['total_renewable_energy'].mean()
print(Avg_gneration)
```

```python
# Maximum generation by state

Max_generation=df.groupby ('state_name')['total_renewable_energy'].max()
print(Max_generation)
```

```python
# SOURCE_WISE TOTAL ENERGY

source_energy=df[['solar_energy','wind_energy','other_renewable_energy']].sum()
print(source_energy)
```

```python
state_summary= df.groupby('state_name').agg({'total_renewable_energy':['sum','mean','max'],
'solar_energy':'sum','wind_energy':'sum'})

print(state_summary)
```

**4,VISULAIZATIONS**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

plt.style.use('ggplot')
```

**1,BAR CHART-STATE-WISE TOTAL RENEWABLE ENERGY GENERATION**

```python
# BAR PLOT - STATE-WISE TOTAL RENEWABLE ENERGY GENERATION

state_generation = df.groupby('state_name')['total_renewable_energy'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,8))
state_generation.plot(kind='bar')

plt.title('STATE-WISE TOTAL RENEWABLE ENERGY GENERATION')
plt.xlabel('state_name')
plt.ylabel('generation (mwh)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# BAR CHART identifies top renewable energy-producing states.
# useful for comparing state performance
```

**2,LINE CHART-DATE WISE RENEWABLE ENERGY TREND**

```python
df['date']=pd.to_datetime(df['date'])
df['year']= df['date'].dt.year

date_generation = df.groupby('year')['total_renewable_energy'].sum()

plt.figure(figsize=(12,6))
plt.plot(date_generation.index,date_generation.values,marker='o')

plt.title('YEAR-WISE RENEWABLE ENERGY GENERATION')
plt.xlabel('year')
plt.ylabel('total renewable energy')
plt.grid(True)

plt.show()
```

```python
# LINE CHART shows growth or decline in renewable energy generation over time
```

**3,PIE CHART-RENEWABLE SOURCE CONTRIBUTION**

```python
source_totals=[df['solar_energy'].sum(),df['wind_energy'].sum(),df['other_renewable_energy'].sum()]

labels=['solar','wind','other']

plt.figure(figsize=(10,7))
plt.pie(source_totals,
        labels= labels,
        autopct= '%1.1f%%')

plt.title('RENEWABLE ENERGY SOURCE CONTRIBUTION')
plt.show()
```

```python
# Pie charts displays percentage contribution of each renewable source
```

**KEY INSIGHTS FROM THE ANALYSIS**




```python
### 1. Renewable Energy Generation Trends
The exploratory analysis revealed noticeable variations in renewable energy generation across states and over time.
Some renewable sources consistently contributed more to total energy generation than others,
indicating differences in resource availability and infrastructure development.

### 2. Distribution of Energy Sources
Histograms and box plots showed that certain energy generation variables were positively skewed,
suggesting that a few states generated significantly higher amounts of renewable energy compared to the majority.

### 3. Correlation Analysis
The correlation heatmap indicated positive relationships between several renewable energy sources and total energy generation.
Strong correlations suggest that increases in individual renewable sources contribute directly to overall renewable energy production.

### 4. State-wise Performance
Groupby and aggregation analyses highlighted states with the highest renewable energy generation.
These states may benefit from favorable geographic conditions, stronger renewable policies,
or greater investment in clean energy infrastructure.

### 5. Outlier Detection
Box plots and kurtosis analysis identified outliers in some energy generation variables.
These extreme values likely represent states with exceptionally high renewable energy production and
should be investigated further for policy and infrastructure insights.
```

```python
## Significant Patterns, Correlations, and Anomalies

- Renewable energy generation is not evenly distributed across all states.
- Wind and solar energy showed substantial growth and contribution in several regions.
- Strong positive correlations were observed between renewable energy sources and total generation.
- A small number of states appeared as outliers due to exceptionally high energy production.
- Some energy sources exhibited high kurtosis values, indicating heavy-tailed distributions and the presence of extreme observations.
- Seasonal or yearly fluctuations may be observed depending on the time period covered by the dataset.
```
