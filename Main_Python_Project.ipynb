```python
import pandas as pd

df=pd.read_csv("https://raw.githubusercontent.com/stephenkumar092-commits/State-Renewable-Energy-Generation/refs/heads/main/state-control-renewable-energy-generation.csv")

print(df.head())
```

```python
# Data types of each column

print("/t Data types:")
print(df.dtypes)
```

```python
# Initial observations (e.g, head(),info(),describe())

# Dataset informartion

print("/Dataset info:")
print(df.info())
```

```python
# Statistical summary

print("/t statistical summary:")
print(df.describe())
```

```python
#2 Data Pre-processing

# check missing values

print("/t missing values:")
print(df.isnull().sum())
```

```python
# REMOVING DUPLICATES

# Check duplicate records
print(df.duplicated().sum())
```

```python
# correcting data types
print(df.dtypes)

# convert date to integer
df['date'] = pd.to_datetime(df['date'])

energy_columns= ['solar_energy','wind_energy','other_renewable_energy','total_renewable_energy']

for col in energy_columns: df[col]=pd.to_numeric(df[col],errors='coerce')
```

```python
#Creating deriverd columns

df['total_renewable_energy']= (
    df['solar_energy']+ df['wind_energy']+ df['other_renewable_energy'])

df['solar_percentage']=(df['solar_energy']/df['total_renewable_energy'])*100

df['wind_percentage']= (df['wind_energy']/df['total_renewable_energy'])*100

df['other_renewable_energy_percentage']= (df['other_renewable_energy']/df['total_renewable_energy'])*100

df['growth_rate']=(df.groupby('state_name')['total_renewable_energy'].pct_change()*100 )

df[[ 'state_name','total_renewable_energy','solar_percentage','wind_percentage','other_renewable_energy_percentage']].head()
```

```python
# filtering and aggregatting data

#Filltering data

# spcific date
filtered_data=df[df['date']== '2020-05-28']
print(filtered_data)
```

```python
#filter specific state

Tamil_Nadu_data= df[df['state_name']=='Tamil Nadu']
print(Tamil_Nadu_data.head())
```

```python
# Aggregatting data

#State wise total Renewable energy generation

state_generation= df.groupby('state_name')['total_renewable_energy'].sum()
print(state_generation)
```

```python
#Average generation by state

Avg_gneration=df.groupby('state_name')['total_renewable_energy'].mean()
print(Avg_gneration)
```

```python
# Maximum generation by state

Max_generation=df.groupby ('state_name')['total_renewable_energy'].max()
print(Max_generation)
```

```python
# SOURCE_WISE TOTAL ENERGY

source_energy=df[['solar_energy','wind_energy','other_renewable_energy']].sum()
print(source_energy)
```

```python
state_summary= df.groupby('state_name').agg({'total_renewable_energy':['sum','mean','max'],
'solar_energy':'sum','wind_energy':'sum'})

print(state_summary)
```

```python
#3 Exploratory Data Analysis

df.describe()

#(i) Univariate Analysis
energy_cols=['solar_energy','wind_energy','other_renewable_energy','total_renewable_energy']
print(df[energy_cols].describe())
```

```python
#State Distribution

print(df['state_name'].value_counts())
```

```python
#date distribution

print(df['date'].value_counts().sort_index())
```

```python
#(ii) Bivariate Analysis

#State_wise

state_avg= df.groupby('state_name')['total_renewable_energy'].mean()

print(state_avg.sort_values(ascending=False))
```

```python
#Date_wise

date_generation= df.groupby('date')['total_renewable_energy'].sum()

print(date_generation)
```

```python
 #solar_energy vs total_renewable_energy

df[['solar_energy','total_renewable_energy']].corr()
```

```python
#(iii) Multivarite analysis

# Groupby analysis

state_name_date = df.groupby(['state_name','date'])['total_renewable_energy'].sum()
print(state_name_date)
```

```python
# miltiple aggregations

summary=df.groupby('state_name').agg({'solar_energy':'sum','wind_energy':'sum','other_renewable_energy':'sum','total_renewable_energy':['sum','mean']})

print(summary)
```

```python
# pivot table

# state vs date
pivot_table= pd.pivot_table (df,
                            values='total_renewable_energy',
                            index= 'state_name',
                            columns= 'date',
                            aggfunc= 'sum')

print(pivot_table)
```

```python
# correlation analysis

corr_matrix= df[['solar_energy','wind_energy','other_renewable_energy','total_renewable_energy']].corr()

print(corr_matrix)
```
